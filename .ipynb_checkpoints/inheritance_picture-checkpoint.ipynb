```python
import numpy as np
import matplotlib.pyplot as plt
from lgca import get_lgca

tmax = 200
lgca= get_lgca(ib=True, geometry='lin', interaction='inheritance', dims = 100, r_b = 0.8, variation = True)
print('maxlabel_init:', lgca.maxlabel_init)
```

```python
lgca.timeevo(timesteps=tmax, record=True)
#print(lgca.props['lab_m'])
lgca.plot_prop_spatial(prop = 'lab_m')
lgca.plot_prop_spatial()
#lgca.plot_density()
```
