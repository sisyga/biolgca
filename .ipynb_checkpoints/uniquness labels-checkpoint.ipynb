```python
from lgca import get_lgca
import numpy as np

lgca= get_lgca(ib=True, geometry='lin', interaction='birth')
```

```python
timesteps = 30
lgca.timeevo(timesteps, record=True)
%matplotlib notebook
lgca.plot_density()
#print(lgca.dens_t.sum(-1)) 
```

```python
#timesteps = 1
uniqueness = False
for i in range(timesteps + 1):
    num = 0
    m = np.concatenate(lgca.nodes_t[i,:,:])
    #l = len(m)
    for j in range(len(m)):
        if m[j] != 0:
            num = num +1
    if num == lgca.dens_t.sum(-1)[i]:
        uniqueness = True
    else:
        uniqueness = False
if uniqueness:
    print('labels are unique!')
else:
    print('labels are NOT unique!')

```
