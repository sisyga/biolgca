```python
from lgca import get_lgca
from lgca.helpers import *
import numpy as np
import matplotlib.pyplot as plt
from matplotlib import cm
%matplotlib notebook

timesteps = 300
dim = 10
dens =  1

#rb = [0.02, 0.1, 0.3, 0.5, 0.7, 0.9, 1]
birthrate = [0.1]
rc = 0
deathrate = 0.02

for i in range(len(birthrate)):
    lgca= get_lgca(ib=True, geometry='lin', interaction='inheritance',\
               density = dens, dims = dim, r_b = birthrate[i], variation = False, restchannels = rc ,r_d = deathrate)
    id = i+1
    lgca.timeevo(timesteps, record=True)
    %matplotlib notebook
    bar_stacked(lgca, save = False, id = id)
    ana_si(lgca, p = True, save = False, id = id)
#     save_data(lgca, id = id)

y = np.arange(timesteps+1)

colors = cm.hsv(y / float(max(y)))
plot = plt.scatter(y, y, c=y, cmap = 'hsv')
plt.clf()
plt.colorbar(plot)
plt.bar(range(len(y)), y, color = colors)
plt.show()
```
