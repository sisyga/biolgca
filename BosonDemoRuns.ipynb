```python
import numpy as np
import matplotlib.pyplot as plt

from lgca import get_lgca
from copy import deepcopy
from operator import itemgetter
%matplotlib notebook
```

```python
# An example initial nodes 
L = 900
k = 3 #number of channels
init_nodes_filled = 10
init_channel_pop = 1
nodes = np.empty(L*k,dtype=object)
for i in range(init_nodes_filled*k):
    nodes[i] = [init_channel_pop*i+j+1 for j in range(init_channel_pop)]
for i in range((L-init_nodes_filled)*k):
    nodes[i+init_nodes_filled*k] = []
newnodes = nodes.reshape(L,k)
newnodes.shape
maxlabel = k*init_nodes_filled*init_channel_pop
```

```python
lgca = get_lgca(geometry = '1d',dims=L, ib=True, ve = True, nodes=newnodes, maxlabel=maxlabel, ini_channel_pop = init_channel_pop, capacity = 8, bc='reflecting', interaction = 'go_or_grow', kappa = 0.0, theta = 0.5, r_b = 0.5, r_d = 0.05)
lgca.timeevo(timesteps = 6000,recordnodespop = True, record=True)
```

```python
lgca.plot_nodes_population()
```

```python
lgca.plot_prop_spatial(propname='kappa')
```

```python
lgca.plot_prop_spatial(propname='theta')
```
