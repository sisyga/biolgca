```python
from lgca import get_lgca
from lgca.helpers import *
import numpy as np
import scipy as sp
import matplotlib.pyplot as plt
# from matplotlib import cm
%matplotlib notebook

timesteps = 300
dim = 5
dens = 0.5

birthrate = 0.8
deathrate = 0.02
rc = 0
m = 1

rep = 100
i = 0
hom_time = np.zeros(rep)
```

```python
while i < rep:
    id = i
    lgca = get_lgca(ib=True, geometry='lin', interaction='inheritance', r_int = m,\
               density = dens, dims = dim, r_b = birthrate, variation = False, restchannels = rc ,r_d = deathrate)
    lgca.timeevo(timesteps, record=True)
    for t in range(timesteps):
        if lgca.sim_ind[t] == 0:
            print('Homogeneity since k = ', t)
            hom_time[i] = t
            break
    i += 1
print(hom_time)
```

```python
# val = sp.stats.itemfreq(hom_time)
val = np.bincount(hom_time.astype(int))

print(len(val))
plt.figure(num=None)
t = 1

while t < len(val):
    plt.bar(t, val[t], 1, color='seagreen')
    t += 1
plt.grid(True)
plt.ylabel('frequency')
plt.xlabel('timepoint of homogeneity')
# plt.yticks(np.arange(0, timesteps, 10))
# plt.xticks(np.arange(0, rep, 1))
plt.tight_layout()
plt.show()
```
