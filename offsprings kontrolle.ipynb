```python
from lgca import get_lgca
from lgca.helpers import *
from lgca.analysis import *
import numpy as np
import os
import matplotlib.pyplot as plt

rc = 0
dim = 90
rep = 500
path = 'pummelzeugs_90/'
file = 'e0797e9'
names = []

for i in range(rep):
    names.append('180' + str(dim) + '_' + str(i) + '_' + file)
for i, name in enumerate(names):
    print('bin bei ', i)
    data = {'offs': np.load('saved_data/' + path + name + '_offsprings.npy'),\
          'nodes': np.load('saved_data/' + path + name + '_nodest.npy'),\
          'labels': np.load('saved_data/' + path + name + '_labels.npy')}

    tmax, dim, ch = data['nodes'].shape
    print(tmax, dim, ch)
    rc = ch - 2
    numbers = np.zeros((tmax, dim*ch + 1))
    numbers[:, 0] = -99


    for t in range(tmax):      
        for d in range(dim):
            for c in range(rc + 2):
                lab = data['nodes'][t,d,c]    
                if lab == 0:
                    continue
                else:
                    lab = data['labels'][lab]
                    numbers[t, lab] += 1

    # print(numbers[11:16, :])

    for t in range(tmax):
        for i in range(0, dim*ch + 1):
            if numbers[t, i] != data['offs'][t,i]:
                print('Fehler bei: ', t, i)
```
