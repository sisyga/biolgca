Vorbereitung:

```python
from lgca import get_lgca
from lgca.helpers import *
from lgca.analysis import *
import numpy as np
import matplotlib.pyplot as plt
import pathlib
```

Daten einlesen und Auswahl treffen:

```python
variation = 'Testdaten'
filename = ''
path = 'saved_data/'

selection = [1,2]
ids = []
for s in selection:
    ids.append(variation + '_' + filename + '_' + str(s))
print(ids)
```

Entropien vergleichen:

```python
for i, s in enumerate(selection):
    data = np.load(path + variation + '_' + str(s) + '_' + filename + '_offsprings.npy')
    plot_index(calc_hillnumbers(data), which='hill2', save=False, id=ids[i])
#     plot_index(calc_shannon(data), which='shannon')
#     plot_hillnumbers_together(hill_1=calc_hillnumbers(data, order=1), hill_2=calc_hillnumbers(data, order=2),\
#                               hill_3=calc_hillnumbers(data, order=3))
#     plot_entropies_together(simpson=calc_simpson(data), shannon=calc_shannon(data), gini=calc_ginisimpson(data))
#     plot_selected_entropies(shannon=calc_shannon(data), gini=calc_ginisimpson(data), hill2=calc_hillnumbers(data))
```

Entwicklung Populationsgröße:

```python
for i, s in enumerate(selection):
    data = np.load(path + variation + '_' + str(s) + '_' + filename + '_offsprings.npy')
    plot_popsize(data, save=False, id=ids[i])
```

Mullerplot erstellen:

```python
for i, s in enumerate(selection):
    data = np.load(path + variation + '_' + str(s) + '_' + filename + '_offsprings.npy')
    mullerplot(data, int_length=5, save=False, id=ids[i])
```
